```python
import pandas as pd
```

```python
df = pd.read_csv("DATASETS/Bengaluru_House_Data.csv")
```

```python
# NULL VALUES
df.isnull()
```

```python
# COUNT BY COLUMN
df.isnull().sum()
```

```python
#TOTAL COUNT
df.isnull().sum().sum()
```

```python
#REPLACE WITH ABOVE COLUMN VALUE
df5 = df.fillna(method='pad')
df5
```

```python
#REPLACE WITH BELOW COLUMN VALUE
df5 = df.fillna(method='bfill')
df5
```

```python
#REPLACE WITH ABOVE COLUMN VALUE PREVIOUS COLUMN
df5 = df.fillna(method='pad', axis=1)
df5
```

```python
#REPLACE WITH ABOVE COLUMN VALUE NEXT COLUMN
df5 = df.fillna(method='bfill', axis=1)
df5
```

```python
#REPLACE WITH ABOVE COLUMN VALUE PREVIOUS COLUMN
df5 = df.fillna(value = df['bath'].mean())
df5
```

```python
#TOTAL NON EMPTY VALUES
df.notnull().sum().sum() - df.isnull().sum().sum()
```
