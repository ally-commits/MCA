```python
import pandas as pd
import numpy as np
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
from sklearn import metrics
```

```python
df = pd.read_csv("DATASETS/mushrooms.csv")
df.head()
```

```python
df = df.apply(LabelEncoder().fit_transform)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
X = df.drop(['class'], axis=1)
Y = df['class']
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3)
arr = []
```

```python
dc = DecisionTreeClassifier(criterion = 'entropy', max_depth=1)
```

```python
AdaBoost = AdaBoostClassifier(base_estimator = dc, n_estimators = 400, learning_rate = 1)
boost_model = AdaBoost.fit(x_train, y_train)
y_pred = boost_model.predict(x_test)

arr.append(metrics.accuracy_score(y_test, y_pred))

metrics.accuracy_score(y_test, y_pred)
```

```python
AdaBoost = AdaBoostClassifier(base_estimator = dc, n_estimators = 200, learning_rate = .002)
boost_model = AdaBoost.fit(x_train, y_train)
y_pred = boost_model.predict(x_test)

arr.append(metrics.accuracy_score(y_test, y_pred))

metrics.accuracy_score(y_test, y_pred)
```

```python
dc = GaussianNB()
```

```python
AdaBoost = AdaBoostClassifier(base_estimator = dc, n_estimators = 400, learning_rate = 1)
boost_model = AdaBoost.fit(x_train, y_train)
y_pred = boost_model.predict(x_test)

arr.append(metrics.accuracy_score(y_test, y_pred))


metrics.accuracy_score(y_test, y_pred)
```

```python
AdaBoost = AdaBoostClassifier(base_estimator = dc, n_estimators = 200, learning_rate = .002)
boost_model = AdaBoost.fit(x_train, y_train)
y_pred = boost_model.predict(x_test)

arr.append(metrics.accuracy_score(y_test, y_pred))


metrics.accuracy_score(y_test, y_pred)
```

```python
dfOutput = pd.DataFrame({
    'Alog': ['Descision tree n_estimators = 400',
             'Descision tree n_estimators = 200',
             'GaussianNB tree n_estimators = 400',
             'GaussianNB tree n_estimators = 200',
            ],
    'Accuracy': arr
})
dfOutput.to_csv("OUTPUT/ensemble_output.csv")
```
