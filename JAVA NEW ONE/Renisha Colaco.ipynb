# Problem statement 1

a) Create three sine functions: sin(x), sin (2x) and sin (x/2), where x=(-pi,+pi) and has a total of 256 values.\
b) Plot these three functions in matplotlib using three different line styles and colours.\
c) Can you overplot sin(x) and cos(x)?\
d) Save the figure as test.png.

```python
import numpy as np
#a
X = np.linspace(-np.pi, np.pi, 256)

sin1 = np.sin(X)
sin2 = np.sin(2*X)
sin3 = np.sin(3*X)
```

```python
#b
import matplotlib.pyplot as plt
plt.title("The plotting of all the functions with the overplotting of cos(x) as well : ")
plt.plot(X, sin1, color="blue", linewidth=1.0, linestyle="-",label='sin(x)')
plt.plot(X, sin2, color="green", linewidth=1.0, linestyle="-",label='sin(2x)')
plt.plot(X, sin3, color="red", linewidth=1.0, linestyle="-",label='sin(x/2)')
```

```python
# c
sin = np.sin(X)
cos = np.cos(X)

plt.plot(X,sin, color='Blue', linewidth=1.0, linestyle="--", label='Sin')
plt.plot(X,cos, color='Green', linewidth=1.0, linestyle="-", label='Cos')

#d
plt.savefig("one.jpg")
```
